```python
import requests

url = "https://raw.githubusercontent.com/Sangameeee/Audio_classifications-for-Age-Gender-Different-Domain-for-Nepali-and-Maithili-Language/refs/heads/main/%20age_gender/age_gender_predictor.py"
response = requests.get(url)
response.raise_for_status()  

with open("age_gender_predictor.py", "w") as f:
    f.write(response.text)
print("✓ Downloaded age_gender_predictor.py")
```

```python
import torch
from age_gender_predictor import AgeGenderPredictor, display_results
device = "cuda" if torch.cuda.is_available() else "cpu"
predictor = AgeGenderPredictor(device='cuda')
```
